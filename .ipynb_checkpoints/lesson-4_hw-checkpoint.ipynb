Решите аналитически и потом численно (в программе) уравнение, зависящее от параметра а: 
sin(а*x)=0
при условии: 0.01<a<0.02, 100<х<500.
Т.е. надо найти решение х как функцию параметра а - построить график x=x(а).
Если численным методом не получается найти все ветви решения x(а), то отыщите хотя бы одну.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Находим значения x и a при которых sin(ax) = 0
a = np.linspace(0.011,0.019,900)
n = 0
x1 = []
a1 = []

for x in range(100,501):
    for el in a:
        if round(x * el, 2) == round(np.pi, 2):
            n += 1
            a1.append(el)
            x1.append(x)
```

```python
# Строим график функции x(a)

plt.figure(figsize=(16, 6))
plt.plot(a1, x1, label=f'x(a)')

title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#808080",
    "family": "serif",
}
plt.title("График функции x(a)", fontdict=title_font, loc="center")

label_font = {
    "fontsize": 12,
    "family": "serif",
}
plt.xlabel("Переменная a", fontdict=label_font)
plt.ylabel("Функция x(a)", fontdict=label_font)

legend_font = {
    "size": 12,
    "family": "serif",
}
plt.legend(loc="upper right", prop=legend_font)

plt.show()
```
