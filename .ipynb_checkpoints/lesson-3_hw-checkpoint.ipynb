```python
def polar_angle(r=0, phi=0):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x, y

r = 10 # Полярный радиус
phi_grad = 90 # Полярный угол (градусов)
phi = phi_grad * np.pi/180 # Перевод градусов в радианы
coords = polar_angle(r, phi)
coords_rad = polar_angle(r, phi_grad)
print(f'В градусах:\nx = {round(coords[0], 2)}\ny = {round(coords[1], 2)}\n{"*" * 30}\nВ радианах:\nx = {round(coords_rad[0], 2)}\ny = {round(coords_rad[1], 2)}')
```

```python
# Вариант 2
# Отрезок строится в Декартовой системе координат через полученые его полярные координаты
# по заданным координатам точек его начала и конца в Декартовой системе координат
a = np.array([3, 5]) # Точка a отрезка ab
b = np.array([8, 1]) # Точка b отрезка ab

x = np.linspace(a[0], b[0], 1000)
y = np.linspace(a[1], b[1], 1000)

def polar_coords(x, y):
    r = np.sqrt(x ** 2 + y ** 2)
    cos_phi = x / np.sqrt(x ** 2 + y ** 2)
    phi_rad = np.arccos(cos_phi)
    
    return r, phi_rad

x1, y1 = polar_coords(x, y)

plt.plot(x1, y1)
plt.grid(axis = 'both')

plt.show()
```

1. Решите систему уравнений

exp(x) + x * (1 – y) = 1  
y = x^2 – 1

```python
# Для наглядности построим график функций
x = np.linspace(-10, 10, 1000)

y1 = x ** 2 - 1
y2 = np.exp(x) / x + 1 - 1 / x

plt.figure(figsize=(8, 12))
plt.plot(x,y1)
plt.plot(x,y2)
plt.ylim(-2,20)
plt.xlim(-4,5)
plt.grid(axis = 'both')

plt.show()
```

```python
# Поиск решений методами библиотеки scipy
# y = x ^ 2 – 1
# y = (e^x - 1) / x + 1

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return ((np.exp(x) - 1 )/ x + 1, x ** 2 - 1)

x1, y1 = fsolve(equations, (-1.58, 1.5))
x1, y1
```

```python
'''
Встроеная функция fsolve как-то криво работает, даже с максимально приближенными
к точке пересечения координатами опорной точки

Для проверки написал свою функцию поиска решения методом перебора.
'''
# y = x ^ 2 – 1
# y = (e^x - 1) / x + 1

def equations(start, stop, step):
    try:
        x = start
        test_range = (abs(start) + abs(stop))  / step
        result = []

        for i in range(int(test_range)):
            x += step
            y1 = x ** 2 - 1
            y2 = (np.exp(x) - 1 )/ x + 1

            if round(y1, 3) == round(y2, 3):
                result.append((round(x, 3), round(y1, 3)))
                
    except Exception as ex:
        print(ex)
            
    return result

start = -4.5 # начельное значение x для перебора. Взято по графику функций приблизительно.
stop = 4.5 # конечное значение x для перебора. Аналогично
step = 0.0001 # шаг перебора
result = equations(start, stop, step)

if result:
    print('Решения системы уравнений:')
    spam = 1
    for i in range(len(result)):
        if i != 0 and result[i][0] != result[i - 1][0]:
            print(f'x{spam}={result[i][0]}, y{spam}={result[i][1]}')
            spam += 1
        elif i == 0:
            print(f'x{spam}={result[i][0]}, y{spam}={result[i][1]}')
            spam += 1
            
else:
    print('Система уравнений не имеет решений.')
```

2. Решите систему неравенств:

y = x^2 – 1

exp(x) + x * (1 – y)  - 1 > 0

```python
'''
y = x ^ 2 – 1
y < (e^x - 1) / x + 1
т.е x ^ 2 – 1 < (e^x - 1) / x + 1
Точки пересечения данных функций мы нашли в предыдущем задании.
По графику видно, что решением будут значения x от 1й точки пересечения до 2й (кроме ОДЗ x!=0, т.к. (e^x - 1) / x нельзя делить на 0)
и от 4й до +∞.
Попробуем это проверить. Буду делать также через свою функцию поиска решения методом перебора.
'''

def equations(start, stop, step):
    try:
        x = start
        test_range = (abs(start) + abs(stop))  / step
        result = []

        for i in range(int(test_range + 1)):
            x += step
            y1 = x ** 2 - 1
            y2 = (np.exp(x) - 1 )/ x + 1

            if round(y1, 3) < round(y2, 3):
                result.append((round(x, 3), round(y1, 3)))
                
    except Exception as ex:
        print(ex)
            
    return result

start = -4.5 # начельное значение x для перебора. Взято по графику функций приблизительно.
stop = 4.5 # конечное значение x для перебора. Аналогично
step = 0.0001 # шаг перебора
result = equations(start, stop, step)

# Вывод результатов

# if result:
#     print('Решения системы уравнений:')
#     spam = 1
#     for i in range(len(result)):
#         if i != 0 and result[i][0] != result[i - 1][0]:
#             print(f'x{spam}={result[i][0]}, y{spam}={result[i][1]}')
#             spam += 1
#         elif i == 0:
#             print(f'x{spam}={result[i][0]}, y{spam}={result[i][1]}')
#             spam += 1
            
# else:
#     print('Система уравнений не имеет решений.')
```
