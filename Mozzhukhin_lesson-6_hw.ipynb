##### Задание 1 (теорема сложения)

Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.

```python
# Вероятность выпадения 2 на кубике = 1/6, вероятность выпадения 5 также = 1/6.
# Нас интересует исход, когда выпадет либо 2, либо 5, поэтому вероятности их выпадения складываем
prob = 1/6 + 1/6
prob
```

##### Задание 2 (теорема умножения)

Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.

```python
# Вероятности выпадения 2 и 5 не изменились. Но теперь нас интересует исход,
# когда в 1м испытании выпадет 2, а затем во 2м 5. Т.е. должны произойти оба события последовательно. Вероятности умножаем
prob2 = 1/6 * 1/6
prob2
```

##### Задание 3

Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости. Обратите внимание на порядок выпадения костей!

```python
# Не совсем понятна формулировка. Если речь идет об одновременном выпадении 2 и 5 в ходе каждого из бросков, то вероятность = 0,
# т.к. это несовместные события. При броске кубика не может выпасть больше одного числа одновременно.

# Если имеется в виду, что в результате первого броска должна выпасть 2, а при втором броске 5,
# то вероятность будет точно такая же, как и в предыдущем задании, т.к. условие существенно не поменялось.
# Нам необходимо, чтобы в результате 1го броска выпала 2, вероятность этого = 1/6, затем, чтобы в результате 2го броска выпала 5,
# вероятность этого также = 1/6, нам необходимо, чтобы произошли оба этих исхода,
# учитывая, то это независимые события, то вероятности также перемножаем.
prob3 = 1/6 * 1/6
prob3
```

#####  Задание 4 (Геометрическая вероятность +интервалы)

На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?

```python
AB = 20
# AC<= 9
# CB <= 15
# Эти 2 условия выполняются на следующем интервале
interval = abs(AB - 15 - 9)
prob4 = interval/AB
prob4
```

#####  Задание 5

Телефонный номер состоит из 7 цифр. Какова вероятность, 
что это номер 8882227?

```python
from math import factorial

# Цифра номера может иметь значение 0-9, т.е. 10 вариантов.
# Каждая цифра может принять любое из значений вне зависимости от того, какое значение приняли остальные цифры.
# По условию нет ограничений на какие-либо цифры номера, т.о. у нас есть 1 возможный вариант данного номера из всех.
# Всего возможных вариантов номеров:
total = 10 ** 7
prob5 = 1 / total
print(prob5)

# Либо можно посчитать вероятность исходя из того, что каждая цифра номера является независимым событием
# и ее вероятность = 1/10, вероятность получения любого номера - это произведение вероятностей получения каждой из его цифр
prob6 = 1/10*1/10*1/10*1/10*1/10*1/10*1/10
print(prob6)
print(round(prob5, 20) == round(prob6, 20))
```

#####  Задание 6

Набирая номер телефона, абонент забыл 2 последние цифры, и, помня только то, что эти цифры различны и среди них нет нуля, стал набирать их наудачу. Сколько вариантов ему надо перебрать, чтобы наверняка найти нужный номер? Какова вероятность того, что он угадает номер с первого раза?

```python
# Вариантов набора 2х цифр номера - 10^2. Мы исключаем варианты, при которых эти цифры могут быть 0, т.о.
# вариантов остается 9^2. Одинаковых сочетаний 2х цифр не много: 0 мы исключили, остается 11, 22, 33, 44, 55, 66, 77, 88, 99
# Итого возможных вариантов:
variants = 9 ** 2 - 9
print(variants)
# Вероятность набрать номер с первого раза равна щансу набрать 1 верную комбинацию из возможных, т.е.
prob7 = 1/variants
print(prob7)
```

#####  Задание 7** (необязательное)

Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью  все грани этого куба будут белыми?

```python
# После окрашивания и разрезания мы получили 27 кубиков из которых:
# 8 кубиков имеют 3/6 окрашеных граней (углы большого куба)
# 12 кубиков имеют 2/6 окрашенных граней (ребра большого куба)
# 6 кубиков имеют 1/6 окрашенную грань (центры граней большого куба)
# Итого мы имеем окрашенных граней кубиков
total_white = 8 * 3 + 12 * 2 + 1 * 6

# Всего граней у полученных 27 кубиков:
total_eges = 27 * 6
print(f'Всего граней: {total_eges}')
print(f'Из них белых: {total_white}')

# Соотношение черных и белых граней не очень большое
print(f'Соотношение: 1 к {total_eges / total_white}')

# Но собрать белый куб возможно только, если грани займут свои изначальные места или
# аналогичные по покраске, как до разбора кубика, а это возможно при таком количестве вариантов
corners = factorial(8) # 8 кубиков могут занять углы и только в 1 позиции
eges = factorial(12) * 2 # 12 кубиков на ребрах и они могут занять на одном и том же месте 2 позиции
centers = factorial(6) # 6 кубиков, которые могут занять центр и только в 1 позиции

# Итого вариантов собрать белый куб:
white_variants = corners * eges * centers
print(f'{"*" * 50}')
print(f'Количество вариантов собрать белый куб: {white_variants}')

# Общее количество вариантов собрать куб:
# Каждый кубик имеет 6 ганей и может занять 1 из 27 позиций, соотвтественно:
total_variants = 6 ** 27
print(f'Количество вариантов собрать куб      : {total_variants}')

# Т.о. вероятность собрать белый куб равна:
print(f'Вероятность собрать белый куб         : {white_variants / total_variants}')
```
