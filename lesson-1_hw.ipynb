Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10,10,1024)

k1 = 1
k2 = 3

plt.figure(figsize=(16, 6))
plt.plot(x, np.cos(k1*x), label=f'y=cos({k1}*x)')
plt.plot(x, np.cos(k2*x), label=f'y=cos({k2}*x)')

title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#808080",
    "family": "serif",
}
plt.title("График функции cos(kx)", fontdict=title_font, loc="center")

label_font = {
    "fontsize": 12,
    "family": "serif",
}
plt.xlabel("Переменная x", fontdict=label_font)
plt.ylabel("Функция cos(kx)", fontdict=label_font)

legend_font = {
    "size": 12,
    "family": "serif",
}
plt.legend(loc="upper left", prop=legend_font)

plt.show()
```
